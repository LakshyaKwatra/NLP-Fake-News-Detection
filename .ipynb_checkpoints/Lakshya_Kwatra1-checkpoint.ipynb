```python
messages = pd.read_csv('train.tsv',sep = '\t',names = ['ID','label','statement','subject','speaker','speaker\'s job title','state','party','barely true counts','false counts', 'half true counts','mostly true counts','pants on fire counts','context'])
```

```python
messages.tail()
```

```python
sns.heatmap(messages.isnull())
```

```python
def text_process(msg):
    npunct = [c for c in msg if c not in string.punctuation]
    npunct = ''.join(npunct)
    return [w for w in npunct.split() if w.lower() not in stopwords.words('english')]
```

```python
text_process(messages['statement'][1045])
```

```python
pipeline = Pipeline([('bow',CountVectorizer(analyzer = text_process)),
                     ('tfidf',TfidfTransformer()),
                     ('classifier',RandomForestClassifier())
                    ])
```

```python
pipeline.fit(messages['statement'],messages['label'])
```

```python
test_messages = pd.read_csv('test.tsv',sep = '\t',names = ['ID','label','statement','subject','speaker','speaker\'s job title','state','party','barely true counts','false counts', 'half true counts','mostly true counts','pants on fire counts','context'])
```

```python
test_messages.tail()
```

```python
pred = pipeline.predict(test_messages['statement'])
```

```python
print(confusion_matrix(test_messages['label'],pred))
```

```python
print(classification_report(test_messages['label'],pred))
```

```python
def makeBinary(label):
    if label in ['false','barely-true','pants-fire']:
        return 'False'
    else:
        return 'True'
```

```python
messages['binary label'] = messages['label'].apply(makeBinary)
```

```python
messages.head()
```

```python
test_messages['binary_label'] = test_messages['label'].apply(makeBinary)
```

```python
test_messages.head()
```

```python
pipeline.fit(messages['statement'],messages['binary label'])
```

```python
pred = pipeline.predict(test_messages['statement'])
```

```python
print(confusion_matrix(test_messages['binary_label'],pred))
```

```python
print(classification_report(test_messages['binary_label'],pred))
```
